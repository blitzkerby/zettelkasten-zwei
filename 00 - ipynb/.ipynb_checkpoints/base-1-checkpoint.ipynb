# Let the scraping commence!

## Steps before scraping a website

1) Fetch the pages (obtained a response object) <br>
`result = requests.get("www.google.com")`

2) Page content <br>
`content = result.text`

3) Create soup <br>
`soup = BeautifulSoup(content, 'lxml')`

## Finding elements with Beautiful Soup

Selecting Single Elements
- `soup.find(id="specific_id")` <br>
- `soup.find('tag', class_="class_name"')` <br>

Selecting Multiple Elements
- `soup.find_all("h2")`


# Scraping Titanic Script

````python
from bs4 import BeautifulSoup as BS
import requests as req

website = "https://subslikescript.com/movie/Titanic-120338"
result  = req.get(website)
content = result.text
soup    = BS(content, 'lxml')

# print(soup.prettify())

box = soup.find("article", class_="main-article")

title = box.find("h1").get_text()
transcript = box.find("div", class_="full-script").get_text(strip=True, separator=" ")

# print(title)
# print(transcript)

with open(f"{title}.txt", 'w') as file:
    file.write(transcript)
````

# Scraping multiple pages

### Update Scripts 2024
In this section, we will learn how to scrape multiple pages with Beautiful Soup. Some of the approaches we will see in this section are workarounds, so don't worry if it seems a bit complicated. In the following sections, I will show you how to do this with few lines of code.

I explain in more detail how waits work in the Selenium section (when I created this course waits weren't necessary on this website)

````python
from bs4 import BeautifulSoup as BU
import requests as req
import time

root = "https://subslikescript.com"

website = f"{root}/movies"
result  = req.get(website)
content = result.text
soup = BU(content, "lxml")
box  = soup.find("article", class_="main-article")

links = []



for link in box.find_all("a", href=True):
    links.append(link["href"])


    
for link in links:
    time.sleep(1)
    
    website = f"{root}/{link}"
    result  = req.get(website)
    content = result.text
    
    soup = BU(content, "lxml")
    
    box = soup.find("article", class_="main-article")
    
    title = box.find("h1").get_text()
    transcript = box.find("div", class_="full-script").get_text(strip=True, separator=" ")

    with open(f"./docs/{title}.txt", "w") as file:
        file.write(transcript)
        
````

# Pagination

````python
import time
import requests as req
from bs4 import BeautifulSoup as bs

# instanciating
root = "https://subslikescript.com/"
page = "movies"

# requesting
request = req.get(root+page)
content = request.text

# making soup
soup = bs(content, "lxml")

# scraping
pagination = soup.find("ul", class_="pagination")

pageDict = {}
pageList = []

for page in pagination.find_all("li", class_="page-item"):
    num = page.text
    ref = page.find("a", href=True)

    try:
        pageDict[int(num)] = ref['href']
    except:
        print(f"Failed to set key {num}")

    


````
